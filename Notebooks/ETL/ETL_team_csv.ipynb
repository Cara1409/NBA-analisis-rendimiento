CARGA ARCHIVO

```python
import pandas as pd
import numpy as np
```

```python
team_csv = 'C:/ProyectoFinalHenry/team.csv'
# Leer el archivo CSV
df = pd.read_csv(team_csv)
print(f"\n✓ Archivo cargado exitosamente")
print(f"✓ Total de registros: {len(df)}")
print(f"✓ Total de columnas: {len(df.columns)}")
```

```python
# Mostrar primeras filas
print(df.head())
```

```python
# Información general del dataset
print("\nInformación del dataset:")
print(df.info())

print("\nEstadísticas descriptivas:")
print(df.describe())

print("\n--- Valores nulos ---")
print(df.isnull().sum())
```

```python
duplicados_antes = len(df)
df = df.drop_duplicates()
duplicados_despues = len(df)
print(f"✓ Duplicados eliminados: {duplicados_antes - duplicados_despues}")
```

```python
# Eliminar espacios en blanco en columnas de texto
columnas_texto = ['full_name', 'abbreviation', 'nickname', 'city', 'state']
for col in columnas_texto:
    if col in df.columns:   
        df[col] = df[col].str.strip()
print(f"✓ Espacios en blanco eliminados en columnas de texto")
```

```python
# Convertir columnas de texto a string (más eficiente)
for col in columnas_texto:
    if col in df.columns:
        df[col] = df[col].astype('string')
```

```python
# Convertir year_founded a int si no hay nulos
if df['year_founded'].isnull().sum() == 0:
    df['year_founded'] = df['year_founded'].astype('int64')
    print("\n✓ year_founded convertido a int64")
else:
    print(f"\n⚠️  year_founded tiene {df['year_founded'].isnull().sum()} valores nulos, se mantiene como float64")

print("\nTipos DESPUÉS de la optimización:")
print(df.dtypes)
```

```python
# Validaciones finales
print("\n--- Validaciones finales ---")
print(f" Total de registros: {len(df)}")
print(f" Total de columnas: {len(df.columns)}")
print(f" Sin duplicados: {df.duplicated().sum() == 0}")
print(f" IDs únicos: {df['id'].nunique() == len(df)}")
print(f" Sin valores nulos: {df.isnull().sum().sum() == 0}")
```

```python
# Guardar datos limpios
archivo_fin = 'team_clean.csv'
df.to_csv(archivo_fin, index=False)
print(f"\n Datos limpios guardados en: {archivo_fin}")
```
